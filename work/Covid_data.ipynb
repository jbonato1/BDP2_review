```python
import requests
import json
import numpy as np
import pandas
import matplotlib.pyplot as plt

```

```python
data_json = requests.get('https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-json/dpc-covid19-ita-andamento-nazionale.json')
data_list = json.loads(data_json.content)
```

```python
tI = []
data=[]
for data_item in data_list:
    tI.append(data_item['terapia_intensiva'])
    data.append(['data'])
```

```python
tI = np.asarray(tI)
```

```python
plt.plot(np.arange(tI.shape[0]),tI)
```
